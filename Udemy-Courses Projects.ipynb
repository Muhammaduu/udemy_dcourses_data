# Q9) What are the courses that are published in 2015?

```python
df.head(2)
```

```python
df.dtypes
```

```python
# first change data type of published_timestamp column into datetime datat type
df['published_timestamp']=pd.to_datetime(df.published_timestamp)
```

```python
df.dtypes
```

```python
# create column for year 
df['year']=df['published_timestamp'].dt.year
```

```python
df.head(2)
```

```python
# now let us the answer the question 
df[df.year==2015]
```
